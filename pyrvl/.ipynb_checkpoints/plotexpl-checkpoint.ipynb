```python
import matplotlib.pyplot as plt
import numpy as np
import m8r

# read data
f = 'bml0.rsf'

inp=m8r.Input(f)
n1=int(inp.get('n1'))
o1=float(inp.get('o1'))
d1=float(inp.get('d1'))
n2=int(inp.get('n2'))
o2=float(inp.get('o2'))
d2=float(inp.get('d2'))
data=inp.read()

# define grid
X, Y = np.meshgrid(np.linspace(o2, o2+(n2-1)*d2, n2), np.linspace(o1, o1+(n1-1)*d1, n1))

#Z=(data.T)[0:200, 0:400]
#print(Z.shape)
#print(X.shape)
#print(Y.shape)

# plot
fig, ax = plt.subplots(figsize=(7,3.5))
pc = ax.pcolormesh(X, Y, (data.T)[0:n1-1,0:n2-1], vmin=np.min(data), vmax=np.max(data), cmap='RdBu_r')
fig.colorbar(pc, ax=ax)
ax.set_title(f)

# y axis increases downwards, seismic-style
plt.ylim(max(plt.ylim()), min(plt.ylim()))
plt.show()
```
